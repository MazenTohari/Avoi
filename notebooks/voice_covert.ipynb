```python
import torch
#import torch_xla
#import torch_xla.core.xla_model as xm
from TTS.api import TTS
import os
import locale
import pandas as pd
from google.colab import drive
```

#Connecting to google drive to access and read csv files of the dataset for cloning

```python
#this command is to remove any google colab checkpoint folder cause they get listed in os.listdir() call
%rm -rf `find -type d -name .ipynb_checkpoints`
device = "cuda" if torch.cuda.is_available() else "cpu"
#dev = xm.xla_device()
drive.mount('/content/drive/', force_remount=True)
path = '/content/drive/MyDrive/voice_project/SADA_SEG'
d_list = os.listdir(path)
df_male = pd.read_csv('/content/drive/MyDrive/voice_project/male.csv')
df_female = pd.read_csv('/content/drive/MyDrive/voice_project/female.csv')
#for cloning we used a pre-triend model provided by MIT called freevc24 it can be used with multiple languages with a small performance requirement
tts = TTS(model_name="voice_conversion_models/multilingual/vctk/freevc24", progress_bar=False).to("cuda")
```

#Cloning process using TTS

```python
for i in d_list:
  print(i)
  #Source audio is our dataset, and the target is the audio file of the person you want to voice clone.
  #We had over 45 minutes of audio in target, because with less audio duration, the worse cloning result would be.
  #Due to GPU RAM limitations, we were unable to go beyond 30 minutes.
  tts.voice_conversion_to_file(source_wav="%s/%s"%(path, i), target_wav="/content/drive/MyDrive/voice_project/HI_30min.wav", file_path="/content/drive/MyDrive/voice_project/converted_dataset/c_%s"%(i))
```

```python
#same segment as the above but over a different audio target
for i,r in df_male.iterrows():
    print(i)
    tts.voice_conversion_to_file(source_wav="%s/c_%s%s"%(path, (str(r['SegmentStart'])+"_"+str(r['SegmentEnd'])), r['FileName'][8:]), target_wav="/content/drive/MyDrive/voice_project/nasser_30min16.wav", file_path="/content/drive/MyDrive/voice_project/converted_dataset/conv_%s%s"%((str(r['SegmentStart'])+"_"+str(r['SegmentEnd'])), r['FileName'][8:]))
```
